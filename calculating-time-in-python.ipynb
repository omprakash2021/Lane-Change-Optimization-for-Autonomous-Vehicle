```python
import time
```

```python
start = time.time()
sum = 0
for i in range(1, 1001):
    sum += i

time.sleep(3) #wait for 3 seconds
print("total time:", time.time() - start)
print(sum)
```

CPU execution time(does not include sleep time)

```python
start = time.process_time()
sum = 0
for i in range(1, 1001):
    sum += i
time.sleep(3)
print("----%s seconds ----" % (time.process_time() - start))
print(sum)
```
